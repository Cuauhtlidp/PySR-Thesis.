```python
import numpy as np
X = 2*np.random.randn(100,5)
y = 2.5382*np.cos(X[:,3])+X[:,0]**2-0.5
```

```python
from pysr import PySRRegressor

model = PySRRegressor(
    maxsize = 20, #Sirve para fijar un límite máximo en el número de nodos (cada nodo es un elemento de la función)
    niterations = 40,
    binary_operators = ["+", "*"],
    unary_operators = [
        "cos",
        "sin",
        "exp",
        "inv(x) = 1/x", #Esta función se define en Julia qué es donde ocurre todo el cálculo.
    ],
    extra_sympy_mappings = {"inv" : lambda x: 1/x}, #Aquí definimos el operador para que Sympy pueda reconocerlo
    elementwise_loss = "loss(prediction, target) = (prediction-target)^2"
)
```

```python
model.fit(X,y)
```

```python
#Lo siguiente me permite ver las predicciones del modelo
model.predict(X)
```

```python
#También es posible realizar predicciones usando un modelo específico de la lista
model.predict(X,0)
```

Algo es importante saber es que PySR guarda el estado interno de la busqueda evolutiva  
de forma que si  nosotros aplicamos de nuevo `model.fit` **PySR** no empieza desde cero  
si no donde se quedó. Puedes desactivar esto al definir el modelo usando `warm_start = True`.  
Si quieres que reinicie el estado puedes usar `model.reset()`.
